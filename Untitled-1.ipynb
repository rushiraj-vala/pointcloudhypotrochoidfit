```python
import numpy as np
import open3d as op3
import matplotlib.pyplot as plt
```

Import the DATA

```python
# Import the data
pcd = op3.io.read_point_cloud("110mm PVC Pipe - front.ply")
points = pcd.points
XYZ = np.asarray(points)
X = XYZ[:,0]
Y = XYZ[:,1]
Z = XYZ[:,2]

# plot the result
figure, axis = plt.subplots(3,1,squeeze=False,figsize=(8,18))

axis[0,0].scatter(X,Y,s=0.01)
axis[0,0].set_title("front")
axis[1,0].scatter(Z,Y,s=0.01)
axis[1,0].set_title("Side")
axis[2,0].scatter(Z,X,s=0.01)
axis[2,0].set_title("Top")

```
